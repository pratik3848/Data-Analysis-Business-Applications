```python
def NormalProbability(ys):
    """Generates data for a normal probability plot.

    ys: sequence of values
  
    returns: numpy arrays xs, ys
    """
    n = len(ys)
    xs = np.random.normal(0, 1, n)
    xs.sort()
    ys = np.array(ys)
    ys.sort()

    return xs, ys
```
